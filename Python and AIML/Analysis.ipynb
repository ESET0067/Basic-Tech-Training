# matplotlib operation

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
fig,ax = plt.subplots()
ax.plot(x, y, marker='o',label='data points')
ax.set_title('Simple Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
plt.plot(x, y)
```

# combined example 

```python
arr = np.array([1, 2, 3, 4, 5])
df = pd.DataFrame({"numbers": arr})
plt.bar(df.index, df["numbers"])
plt.xlabel("Index")
plt.ylabel("Numbers")
plt.title("Bar Plot of Numbers")
plt.show()
```

```python
df1 = np.array([20,20,10,25,25])
plt.pie(df1, labels=['A', 'B', 'C', 'D', 'E'], autopct='%1.1f%%')
plt.title("Pie Chart Example")
```
